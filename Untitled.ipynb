```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df=df.drop("Unnamed: 0",axis=1)
```

# Gender Distribution Bar Graph

```python
ax=sns.countplot(data=df,x="Gender")
plt.figure(figsize=(5,5))
ax.bar_label(ax.containers[0])
plt.show()
```

# Based on the chart analysis, it is evident that the dataset contains a higher representation of females compared to males.

# After examining the chart, we can conclude that there are more female data points than male data points, while also taking outliers into consideration.

```python
group=df.groupby("ParentEduc").agg({"MathScore":'mean',"WritingScore":'mean',"WritingScore":'mean'})
plt.title("Relationship between Parent's Eduction and Student's Score")
sns.heatmap(group,annot=True)
```

# Upon reviewing the chart,we found that parents' educational level has a significant impact on their children's academic ability. In particular, it is clear that highly educated parents  tend to have a positive impact on their children's academic performance.

# Distribution of Ethnic Groups

```python
groupA = df[df["EthnicGroup"] == 'group A'].count()["EthnicGroup"]
groupB = df[df["EthnicGroup"] == 'group B'].count()["EthnicGroup"]
groupC = df[df["EthnicGroup"] == 'group C'].count()["EthnicGroup"]
groupD = df[df["EthnicGroup"] == 'group D'].count()["EthnicGroup"]
groupE = df[df["EthnicGroup"] == 'group E'].count()["EthnicGroup"]
mylist = [groupA, groupB, groupC, groupD, groupE]
l=["group A","group B","group C","group D","group E"]
plt.title("Distribution of Groups Ethnic")
plt.pie(mylist,labels=l,autopct="%1.2f%%")
plt.show()
```
